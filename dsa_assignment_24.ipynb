```python
def romanToInt(s):
    roman_values = {
        'I': 1,
        'V': 5,
        'X': 10,
        'L': 50,
        'C': 100,
        'D': 500,
        'M': 1000
    }
    
    valid_characters = set('IVXLCDM')
    if not 1 <= len(s) <= 15 or set(s) - valid_characters:
        raise ValueError("Invalid input: Roman numeral contains invalid characters.")
    
    result = 0
    n = len(s)
    
    for i in range(n):
        if i < n - 1 and roman_values[s[i]] < roman_values[s[i+1]]:
            result -= roman_values[s[i]]
        else:
            result += roman_values[s[i]]
    
    if result < 1 or result > 3999:
        return 'not in range'
    
    return result

# Test cases
print(romanToInt("III"))     # Output: 3
print(romanToInt("LVIII"))   # Output: 58
print(romanToInt("MMMMCMXCIX")) # not range constraint
```

```python
def lengthOfLongestSubstring(s):
    valid_characters = set(
        "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789!@#$%^&*()-_+=[]{};:,.<>/?~` "
    )
    
    if not (
        0 <= len(s) <= 50000 and
        all(char in valid_characters for char in s)
    ):
        return  "Constraint limit"
    
    start = 0
    end = 0
    max_length = 0
    visited = set()
    
    while end < len(s):
        if s[end] in visited:
            visited.remove(s[start])
            start += 1
        else:
            visited.add(s[end])
            max_length = max(max_length, end - start + 1)
            end += 1
    
    return max_length

# Test cases
print(lengthOfLongestSubstring("abcabcbb"))  # Output: 3
print(lengthOfLongestSubstring("bbbbb"))     # Output: 1
print(lengthOfLongestSubstring("pwwkew"))    # Output: 3
print(lengthOfLongestSubstring(""))          # Output: 0
print(lengthOfLongestSubstring("abc123!@# "))# Output: 10
print(lengthOfLongestSubstring("abc123!@#ðŸ™‚"))# Output: constriant limit
```

```python
def majorityElement(nums):
    n = len(nums)
    
    if not (1 <= n <= 5 * 10**4):
        return " not in the range of the constraints"
    
    for num in nums:
        if not (-10**9 <= num <= 10**9):
            return " not in the range of num"
    
    candidate = None
    count = 0
    
    for num in nums:
        if count == 0:
            candidate = num
            count += 1
        elif num == candidate:
            count += 1
        else:
            count -= 1
    
    return candidate

# Test cases
print(majorityElement([3, 2, 3]))              # Output: 3
print(majorityElement([2, 2, 1, 1, 1, 2, 2]))  # Output: 2
print(majorityElement([]))                     # Output: not in range
print(majorityElement([1] * (5 * 10**4 + 1)))  # Output: not in the range
```

```python
from collections import defaultdict

def groupAnagrams(strs):
    n = len(strs)
    
    if not 1 <= n <= 10000:
        return 'not in range 1 to 1000'
    
    for s in strs:
        if not (0 <= len(s) <= 100) or not all(c.islower() for c in s):
            return 'not in the range 0 to 100'
    
    anagram_groups = defaultdict(list)
    
    for s in strs:
        sorted_s = ''.join(sorted(s))
        anagram_groups[sorted_s].append(s)
    
    return list(anagram_groups.values())

# Test cases
print(groupAnagrams(["eat","tea","tan","ate","nat","bat"]))
# Output: [["eat","tea","ate"],["tan","nat"],["bat"]]

print(groupAnagrams([""]))
# Output: [[""]]

print(groupAnagrams(["a"]))
# Output: [["a"]]

print(groupAnagrams(["apple"] * 10001))
# Output: 

print(groupAnagrams(["abcd", "efghijklmnopqrstuvwxyz" * 10]))
# Output: range constraint

print(groupAnagrams(["abc", "def", "Ghi"]))
# Output: range constraint
```

```python
def nthUglyNumber(n):
    if not 1 <= n <= 1690:
        return ' not in the range of 1 to 1690'
    
    ugly = [0] * n
    ugly[0] = 1
    
    p2 = p3 = p5 = 0
    next_multiple_2 = 2
    next_multiple_3 = 3
    next_multiple_5 = 5
    
    for i in range(1, n):
        next_ugly = min(next_multiple_2, next_multiple_3, next_multiple_5)
        ugly[i] = next_ugly
        
        if next_ugly == next_multiple_2:
            p2 += 1
            next_multiple_2 = ugly[p2] * 2
        
        if next_ugly == next_multiple_3:
            p3 += 1
            next_multiple_3 = ugly[p3] * 3
        
        if next_ugly == next_multiple_5:
            p5 += 1
            next_multiple_5 = ugly[p5] * 5
    
    return ugly[-1]

# Test cases
print(nthUglyNumber(10))        # Output: 12
print(nthUglyNumber(1))         # Output: 1
print(nthUglyNumber(1690))      # Output: 2123366400
print(nthUglyNumber(0))         # Output: ValueError
print(nthUglyNumber(1691))      # Output: ValueError
print(nthUglyNumber(10000))     # Output: ValueError
```

```python
from collections import Counter

def topKFrequent(words, k):
    if not 1 <= len(words) <= 500:
        return 'not in range of 0 to 500'
    
    for word in words:
        if not (1 <= len(word) <= 10) or not word.islower():
            return 'not in the 1 to 10'
    
    counter = Counter(words)
    
    if not 1 <= k <= len(counter):
        return 'not in the range of 1 and length of counter'
    
    sorted_words = sorted(counter.keys(), key=lambda w: (-counter[w], w))
    
    return sorted_words[:k]

# Test cases
print(topKFrequent(["i","love","leetcode","i","love","coding"], 2))
# Output: ["i","love"]

print(topKFrequent(["the","day","is","sunny","the","the","the","sunny","is","is"], 4))
# Output: ["the","is","sunny","day"]

print(topKFrequent([], 3))
# Output: []

print(topKFrequent(["apple"] * 501, 5))
# Output: 

print(topKFrequent(["ab", "abcde", "fghi"], 3))
# Output: 

print(topKFrequent(["abc", "def", "Ghi"], 3))
# Output: 

print(topKFrequent(["abc", "def", "ghi"], 5))
# Output: 
```

```python
from collections import deque

def maxSlidingWindow(nums, k):
    if not 1 <= len(nums) <= 100000 or not 1 <= k <= len(nums):
        return 'not in the range of 1 to 100000'
    
    for num in nums:
        if not -10000 <= num <= 10000:
            return 'not in the range of -10000 to 10000'
    
    result = []
    window = deque()
    
    for i in range(len(nums)):
        # Remove indices from the right end of the window whose corresponding elements are smaller than or equal to nums[i]
        while window and nums[window[-1]] <= nums[i]:
            window.pop()
        
        # Add the current index to the right end of the window
        window.append(i)
        
        # Remove the leftmost index from the window if it's outside the current window position
        if window[0] <= i - k:
            window.popleft()
        
        # Append the maximum element to the result if the window is fully formed
        if i >= k - 1:
            result.append(nums[window[0]])
    
    return result

# Test cases
print(maxSlidingWindow([1,3,-1,-3,5,3,6,7], 3))
# Output: [3,3,5,5,6,7]

print(maxSlidingWindow([1], 1))
# Output: [1]

print(maxSlidingWindow([], 3))
# Output: 

print(maxSlidingWindow([1] * 100001, 5))
# Output: 

print(maxSlidingWindow([-10000, 10000, 5000, -2000, 8000], 2))
# Output: [10000, 10000, 5000, 5000]

print(maxSlidingWindow([1, 2, 3], 4))
# Output: 'not in the range of 100000'
```

```python
def findClosestElements(arr, k, x):
    if not 1 <= len(arr) <= 10000:
        raise ValueError("Invalid input: Length of arr is out of range.")
    
    if not 1 <= k <= len(arr):
        raise ValueError("Invalid input: k is out of range.")
    
    if not isSorted(arr):
        raise ValueError("Invalid input: arr is not sorted in ascending order.")
    
    if not (-10000 <= x <= 10000):
        raise ValueError("Invalid input: x is out of range.")
    
    for num in arr:
        if not (-10000 <= num <= 10000):
            raise ValueError("Invalid input: arr contains elements out of range.")
    
    left = binarySearch(arr, x)
    right = left + 1
    
    while k > 0:
        if left >= 0 and right < len(arr):
            if abs(arr[left] - x) <= abs(arr[right] - x):
                left -= 1
            else:
                right += 1
        elif left >= 0:
            left -= 1
        else:
            right += 1
        
        k -= 1
    
    return arr[left+1:right]

def binarySearch(arr, x):
    low = 0
    high = len(arr) - 1
    
    while low < high:
        mid = (low + high) // 2
        if arr[mid] < x:
            low = mid + 1
        else:
            high = mid
    
    return low

def isSorted(arr):
    for i in range(1, len(arr)):
        if arr[i] < arr[i-1]:
            return False
    return True

# Test cases
print(findClosestElements([1,2,3,4,5], 4, 3))
# Output: [1, 2, 3, 4]

print(findClosestElements([1,2,3,4,5], 4, -1))
# Output: [1, 2, 3, 4]
```
