```python
#2 A number N is represented in Linked List such that each digit corresponds to a node in linked list. You need to add 1 to it.

class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None


def reverseLinkedList(head):
    if not head or not head.next:
        return head

    prev_node = None
    current = head

    while current:
        next_node = current.next
        current.next = prev_node
        prev_node = current
        current = next_node

    return prev_node


def addOne(head):
    # Reverse the linked list
    reversed_head = reverseLinkedList(head)

    current = reversed_head
    carry = 1

    while current:
        sum = current.data + carry
        if sum < 10:
            current.data = sum
            carry = 0
        else:
            current.data = sum % 10
            carry = 1

        current = current.next

    if carry == 1:
        # If there is still a carry, create a new node with the carry as the data
        new_node = Node(carry)
        current = reversed_head

        while current.next:
            current = current.next

        current.next = new_node

    # Reverse the linked list again
    updated_head = reverseLinkedList(reversed_head)

    return updated_head


# Helper function to create a linked list from a given number
def createLinkedList(number):
    if number == 0:
        return Node(0)

    head = None

    while number > 0:
        digit = number % 10
        new_node = Node(digit)

        if not head:
            head = new_node
        else:
            new_node.next = head
            head = new_node

        number //= 10

    return head


# Helper function to print the linked list as a number
def printLinkedListAsNumber(head):
    current = head
    number = 0

    while current:
        number = number * 10 + current.data
        current = current.next

    print(number)


# Test case 1
# Create the linked list: 4 -> 5 -> 6
head1 = createLinkedList(456)

print("Input:")
printLinkedListAsNumber(head1)

# Add 1 to the linked list
updated_head1 = addOne(head1)

print("\nOutput:")
printLinkedListAsNumber(updated_head1)


# Test case 2
# Create the linked list: 1 -> 2 -> 3
head2 = createLinkedList(123)

print("\nInput:")
printLinkedListAsNumber(head2)

# Add 1 to the linked list
updated_head2 = addOne(head2)

print("\nOutput:")
printLinkedListAsNumber(updated_head2)
```

```python
#3 Given a Linked List of size N, where every node represents a sub-linked-list and contains two pointers:(i) a next pointer to the next node,(ii) a bottom pointer to a linked list where this node is head.Each of the sub-linked-list is in sorted order.Flatten the Link List such that all the nodes appear in a single level while maintaining the sorted order. Note: The flattened list will be printed using the bottom pointer instead of next pointer.

class ListNode:
    def __init__(self, val=None, next=None, bottom=None):
        self.val = val
        self.next = next
        self.bottom = bottom


def mergeLists(list1, list2):
    if not list1:
        return list2
    if not list2:
        return list1

    if list1.val <= list2.val:
        list1.bottom = mergeLists(list1.bottom, list2)
        return list1
    else:
        list2.bottom = mergeLists(list1, list2.bottom)
        return list2


def flatten(head):
    if not head or not head.next:
        return head

    head.next = flatten(head.next)
    head = mergeLists(head, head.next)

    return head


# Helper function to create a linked list from a given list of values
def createLinkedList(values):
    if not values:
        return None

    nodes = [ListNode(val) for val in values]

    for i in range(len(nodes) - 1):
        nodes[i].next = nodes[i + 1]

    return nodes[0]


# Helper function to convert a linked list to a list of values using the bottom pointer
def convertToList(head):
    values = []
    current = head
    while current:
        values.append(current.val)
        current = current.bottom
    return values


# Test case 1
values1 = [5, 10, 19, 28]
head1 = createLinkedList(values1)
head1.bottom = createLinkedList([7, 20, 22, 35])
head1.bottom.bottom = createLinkedList([8, 50, 40])
head1.bottom.bottom.bottom = createLinkedList([30, 45])
print("Input:")
current = head1
while current:
    print(current.val, end=" -> ")
    current = current.bottom
print("None")
flattened_head1 = flatten(head1)
flattened_list1 = convertToList(flattened_head1)
print("Output:", flattened_list1)

# Test case 2
values2 = [5, 10, 19, 28]
head2 = createLinkedList(values2)
head2.bottom = createLinkedList([7, 22])
head2.bottom.bottom = createLinkedList([8, 50])
head2.bottom.bottom.bottom = createLinkedList([30])
print("\nInput:")
current = head2
while current:
    print(current.val, end=" -> ")
    current = current.bottom
print("None")
flattened_head2 = flatten(head2)
flattened_list2 = convertToList(flattened_head2)
print("Output:", flattened_list2)
```

```python
#4 You are given a special linked list with **N** nodes where each node has a next pointer pointing to its next node. You are also given **M** random pointers, where you will be given **M** number of pairs denoting two nodes **a** and **b**  **i.e. a->arb = b** (arb is pointer to random node)**.**Construct a copy of the given list. The copy should consist of exactly **N** new nodes, where each new node has its value set to the value of its corresponding original node. Both the next and random pointer of the new nodes should point to new nodes in the copied list such that the pointers in the original list and copied list represent the same list state. None of the pointers in the new list should point to nodes in the original list.For example, if there are two nodes **X** and **Y** in the original list, where **X.arb** **-->** **Y**, then for the corresponding two nodes **x** and **y** in the copied list, **x.arb --> y.**Return the head of the copied linked list.

class Node:
    def __init__(self, val=0, next=None, random=None):
        self.val = val
        self.next = next
        self.random = random


def copyRandomList(head):
    if not head:
        return None

    # Step 1: Create new nodes and insert them after the original nodes
    current = head
    while current:
        new_node = Node(current.val)
        new_node.next = current.next
        current.next = new_node
        current = new_node.next

    # Step 2: Assign random pointers to the new nodes
    current = head
    while current:
        if current.random:
            current.next.random = current.random.next
        current = current.next.next

    # Step 3: Separate the original and copied lists
    new_head = head.next
    current = head
    while current:
        copied_node = current.next
        current.next = copied_node.next
        if copied_node.next:
            copied_node.next = copied_node.next.next
        current = current.next

    return new_head


# Helper function to create a linked list from a given list of values
def createLinkedList(values, pairs):
    if not values:
        return None

    nodes = [Node(val) for val in values]

    for i, (src, dest) in enumerate(pairs):
        nodes[i].random = nodes[dest - 1]

    for i in range(len(nodes) - 1):
        nodes[i].next = nodes[i + 1]

    return nodes[0]


# Test case 1
values1 = [1, 2, 3, 4]
pairs1 = [[1, 2], [2, 4]]
head1 = createLinkedList(values1, pairs1)
print("Input:")
current1 = head1
while current1:
    print(f"Node {current1.val} -> Random: {current1.random.val if current1.random else None}")
    current1 = current1.next

copied_head1 = copyRandomList(head1)

print("\nOutput:")
current1 = copied_head1
while current1:
    print(f"Node {current1.val} -> Random: {current1.random.val if current1.random else None}")
    current1 = current1.next

# Test case 2
values2 = [1, 3, 5, 9]
pairs2 = [[1, 1], [3, 4]]
head2 = createLinkedList(values2, pairs2)
print("\nInput:")
current2 = head2
while current2:
    print(f"Node {current2.val} -> Random: {current2.random.val if current2.random else None}")
    current2 = current2.next

copied_head2 = copyRandomList(head2)

print("\nOutput:")
current2 = copied_head2
while current2:
    print(f"Node {current2.val} -> Random: {current2.random.val if current2.random else None}")
    current2 = current2.next
```

```python
#5 Given the `head` of a singly linked list, group all the nodes with odd indices together followed by the nodes with even indices, and return *the reordered list*.

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def oddEvenList(head):
    if not head or not head.next:
        return head

    odd_head = odd = head
    even_head = even = head.next

    while even and even.next:
        odd.next = even.next
        odd = odd.next
        even.next = odd.next
        even = even.next

    odd.next = even_head

    return odd_head


# Helper function to create a linked list from a given list of values
def createLinkedList(values):
    if not values:
        return None

    nodes = [ListNode(val) for val in values]

    for i in range(len(nodes) - 1):
        nodes[i].next = nodes[i + 1]

    return nodes[0]


# Helper function to convert a linked list to a list of values
def convertToList(head):
    values = []
    current = head
    while current:
        values.append(current.val)
        current = current.next
    return values


# Test case 1
values1 = [1, 2, 3, 4, 5]
head1 = createLinkedList(values1)
print("Input:", values1)
reordered_head1 = oddEvenList(head1)
reordered_list1 = convertToList(reordered_head1)
print("Output:", reordered_list1)

# Test case 2
values2 = [2, 1, 3, 5, 6, 4, 7]
head2 = createLinkedList(values2)
print("\nInput:", values2)
reordered_head2 = oddEvenList(head2)
reordered_list2 = convertToList(reordered_head2)
print("Output:", reordered_list2)
```

```python
#6 Given a singly linked list of size N. The task is to left-shift the linked list by k nodes, where k is a given positive integer smaller than or equal to length of the linked list.

class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None


def leftShiftLinkedList(head, k):
    if not head or not head.next or k == 0:
        return head

    length = 0
    current = head

    # Find the length of the linked list
    while current:
        length += 1
        current = current.next

    if k == length:
        # No need to rotate the list if k is equal to the length of the linked list
        return head

    k = k % length  # Adjust k if it is greater than the length of the linked list

    if k == 0:
        # No need to rotate the list if k becomes 0 after adjustment
        return head

    # Find the kth node from the beginning of the linked list
    current = head
    for _ in range(k - 1):
        current = current.next

    new_head = current.next  # Head of the sublist after rotation

    # Set the next pointer of the kth node to None
    current.next = None

    # Traverse to the end of the original linked list
    current = new_head
    while current.next:
        current = current.next

    # Connect the end of the original linked list to the head of the sublist
    current.next = head

    # Update the head of the linked list to the (k+1)th node
    head = new_head

    return head


# Helper function to create a linked list from a given list of values
def createLinkedList(values):
    if not values:
        return None

    head = Node(values[0])
    current = head

    for value in values[1:]:
        node = Node(value)
        current.next = node
        current = node

    return head


# Helper function to print the linked list
def printLinkedList(head):
    current = head

    while current:
        print(current.data, end=" ")
        current = current.next

    print()


# Test case
# Create the linked list: 2 -> 4 -> 7 -> 8 -> 9
values = [2, 4, 7, 8, 9]
k = 3

head = createLinkedList(values)

print("Input:")
printLinkedList(head)

# Left-shift the linked list by k nodes
updated_head = leftShiftLinkedList(head, k)

print("\nOutput:")
printLinkedList(updated_head)

```

```python
#7 You are given the `head` of a linked list with `n` nodes.For each node in the list, find the value of the **next greater node**. That is, for each node, find the value of the first node that is next to it and has a **strictly larger** value than it. Return an integer array `answer` where `answer[i]` is the value of the next greater node of the `ith` node (**1-indexed**). If the `ith` node does not have a next greater node, set `answer[i] = 0`.

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def nextLargerNodes(head):
    stack = []
    result = []
    index = 0

    # Traverse the linked list
    while head:
        result.append(0)  # Initialize the answer for the current node as 0

        # Check if the current node is greater than the nodes in the stack
        while stack and stack[-1][0] < head.val:
            _, stack_index = stack.pop()
            result[stack_index] = head.val

        stack.append((head.val, index))
        index += 1
        head = head.next

    return result


# Helper function to create a linked list from a given list of values
def createLinkedList(values):
    if not values:
        return None

    head = ListNode(values[0])
    current = head

    for value in values[1:]:
        node = ListNode(value)
        current.next = node
        current = node

    return head


# Test cases
head1 = createLinkedList([2, 1, 5])
print("Input:", [2,1,5])
result1 = nextLargerNodes(head1)
print("Output:", result1)

head2 = createLinkedList([2, 7, 4, 3, 5])
print("\nInput:", [2,7,4,3,5])
result2 = nextLargerNodes(head2)
print("Output:", result2)
```

```python
#8  Given the `head` of a linked list, we repeatedly delete consecutive sequences of nodes that sum to `0` until there are no such sequences.After doing so, return the head of the final linked list.  You may return any such answer.(Note that in the examples below, all sequences are serializations of `ListNode` objects.)

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def removeZeroSumSublists(head):
    dummy = ListNode(0)  # Dummy node to handle the case where the head needs to be removed
    dummy.next = head
    prefix_sum = 0
    prefix_sums = {}
    curr = dummy

    while curr:
        prefix_sum += curr.val

        if prefix_sum in prefix_sums:
            # Remove nodes between prefix_sums[prefix_sum] and curr (inclusive)
            node = prefix_sums[prefix_sum].next
            prefix_sum += node.val
            while node != curr:
                del prefix_sums[prefix_sum]
                node = node.next
                prefix_sum += node.val

            # Update the next pointer of the previous node
            prefix_sums[prefix_sum].next = curr.next
        else:
            prefix_sums[prefix_sum] = curr

        curr = curr.next

    return dummy.next


# Helper function to create a linked list from a given list of values
def createLinkedList(values):
    if not values:
        return None

    head = ListNode(values[0])
    current = head

    for value in values[1:]:
        node = ListNode(value)
        current.next = node
        current = node

    return head


# Helper function to convert a linked list to a list
def linkedListToList(head):
    result = []
    current = head

    while current:
        result.append(current.val)
        current = current.next

    return result


# Test cases
head1 = createLinkedList([1, 2, -3, 3, 1])
print("Input:", linkedListToList(head1))
result1 = removeZeroSumSublists(head1)
print("Output:", linkedListToList(result1))

head2 = createLinkedList([1, 2, 3, -3, 4])
print("\nInput:", linkedListToList(head2))
result2 = removeZeroSumSublists(head2)
print("Output:", linkedListToList(result2))

head3 = createLinkedList([1, 2, 3, -3, -2])
print("\nInput:", linkedListToList(head3))
result3 = removeZeroSumSublists(head3)
print("Output:", linkedListToList(result3))


  
```
