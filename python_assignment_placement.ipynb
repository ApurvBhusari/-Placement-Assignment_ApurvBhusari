```python
#Question 2: - Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if he can remove just one character at the index in the string, and the remaining characters will occur the same number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO .

from collections import Counter

def valid_check(s):
    char_counts = Counter(s)
    counts = list(char_counts.values())
    
    # If all characters have the same count, the string is valid
    if len(set(counts)) == 1:
        return "YES"
    
    
    # If there are exactly two different counts
    elif len(set(counts)) == 2:
        # Check if removing one character makes it valid
        # Find the counts and characters associated with each count
        count1, count2 = set(counts)
        char1 = counts.count(count1)
        char2 = counts.count(count2)
        print(count1,'count1')
        print(count2,'count2')
        print(char1,'char1')
        print(char2,'char2')
        #special condition for those when the magnitude is equal to either is 1 and 2
        if (count1==2 and count2==1 and char2>=2 ) or (count1==1 and count2==2 and char1>=2):
            return 'NO'
        
        # If there is only one character with a count of 1
        elif char1 == 1 and (count1 == 1 or count1 - 1 == count2):
            return "YES"
        
        # If there is only one character with a count of 1
        elif char2 == 1 and (count2 == 1 or count2 - 1 == count1):
            return "YES"
    else:
        return "NO"

# testcase1
string = "abc"
valid = valid_check(string)
print(valid)

#testcase2
string1 ='abcc'
valid1=valid_check(string1)
print(valid1)

#testcase3
string2='aabbccc'
valid2=valid_check(string2)
print(valid2)

#testcase4 
string3='aabbc'
valid3=valid_check(string3)
print(valid3)

#testcase5
string4='abbcccdddd'
valid4=valid_check(string4)
print(valid4)

            
        
```

```python
#Question 10 - Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or paragraph, and return their respective count as a dictionary. Note - 1. Write code comments wherever required for code 2. You have to write at least 2 additional test cases in which your program will run successfully and provide an explanation for the same.

from textblob import TextBlob

def partsofspeech(text):
    blob = TextBlob(text)
    
    # Initialize counters
    a = {'verbs': 0, 'nouns': 0, 'pronouns': 0, 'adjectives': 0}
    
    # Count the POS tags
    for word, tag in blob.tags:
        # Verbs
        if tag.startswith('VB'):
            
            a['verbs'] += 1
            
        # Nouns
        elif tag.startswith('NN'):
            
            a['nouns'] += 1
        # Pronouns
        elif tag.startswith('PRP'):
            
            a['pronouns'] += 1
        # Adjectives
        elif tag.startswith('JJ'):
            
            a['adjectives'] += 1
    
    return a

#1st testcase
text = "I think that the important thing is this world is to believe yourself no matter what the situation is"
pos_counts = partsofspeech(text)
print(pos_counts)
#2nd testcase
text='Your greatness is not defined by how you behave when you are climbing the ladder of success , it is defined when you have everything'
pos_counts1 = partsofspeech(text)
print(pos_counts1)


    
```
