```python
#Given an array arr[ ] of size N having elements, the task is to find the next greater element for each element of the array in order of their appearance in the array.Next greater element of an element in the array is the nearest element on the right which is greater than the current element.If there does not exist next greater of current element, then next greater element for current element is -1. For example, next greater of the last element is always -1.

def find_next_greater_elements(arr):
    stack = []
    result = [-1] * len(arr)  # Initialize result array with -1

    for i in range(len(arr)-1, -1, -1):
        # While the stack is not empty and the current element is greater than or equal to the element at the top of the stack
        while stack and arr[i] >= arr[stack[-1]]:
            stack.pop()  # Pop the element from the stack

        if stack:
            result[i] = arr[stack[-1]]  # Set the next greater element for the current element

        stack.append(i)  # Push the current element's index onto the stack

    return result
# Example 1
arr1 = [1, 3, 2, 4]
next_greater1 = find_next_greater_elements(arr1)
print(next_greater1)  # Output: [3, 4, 4, -1]

# Example 2
arr2 = [6, 8, 0, 1, 3]
next_greater2 = find_next_greater_elements(arr2)
print(next_greater2)  # Output: [8, -1, 1, 3, -1]
```

```python
#Given an array a of integers of length n, find the nearest smaller number for every element such that the smaller element is on left side.If no small element present on the left print -1.

def find_nearest_smaller_elements(a):
    stack = []
    result = []

    for num in a:
        # While the stack is not empty and the top element is greater than or equal to the current number
        while stack and stack[-1] >= num:
            stack.pop()

        if not stack:
            result.append(-1)  # No smaller element found on the left
        else:
            result.append(stack[-1])  # Found the nearest smaller element on the left

        stack.append(num)  # Push the current number onto the stack

    return result
# Example 1
a1 = [1, 6, 2]
nearest_smaller1 = find_nearest_smaller_elements(a1)
print(nearest_smaller1)  # Output: [-1, 1, 1]

# Example 2
a2 = [1, 5, 0, 3, 4, 5]
nearest_smaller2 = find_nearest_smaller_elements(a2)
print(nearest_smaller2)  # Output: [-1, 1, -1, 0, 3, 4]
```

```python
#Implement a Stack using two queues q1 and q2.

class Stack:
    def __init__(self):
        self.q1 = []
        self.q2 = []

    def push(self, value):
        self.q2.append(value)

        # Move elements from q1 to q2
        while self.q1:
            self.q2.append(self.q1.pop(0))

        # Swap q1 and q2
        self.q1, self.q2 = self.q2, self.q1

    def pop(self):
        if not self.q1:
            return -1  # Stack is empty

        return self.q1.pop(0)

# Example usage:
stack = Stack()

stack.push(2)
stack.push(3)
print(stack.pop())  # Output: 3
stack.push(4)
print(stack.pop())  # Output: 4

# Additional example
stack.push(2)
print(stack.pop())  # Output: 2
print(stack.pop())  # Output: -1 (stack is empty)



   
```

```python
#You are given a stack St. You have to reverse the stack using recursion.

class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

    def reverse(self):
        if not self.is_empty():
            # Remove the top element from the stack
            top_element = self.pop()

            # Reverse the remaining stack
            self.reverse()

            # Insert the top element at the bottom of the reversed stack
            self.insert_at_bottom(top_element)

    def insert_at_bottom(self, item):
        if self.is_empty():
            self.push(item)
        else:
            # Remove all elements from the stack and insert them at the bottom recursively
            top_element = self.pop()
            self.insert_at_bottom(item)
            self.push(top_element)

# Example usage:
stack = Stack()

# Example 1
stack.push(3)
stack.push(2)
stack.push(1)
stack.push(7)
stack.push(6)

print("Original Stack:", stack.items)
stack.reverse()
print("Reversed Stack:", stack.items)

# Example 2
stack = Stack()
stack.push(4)
stack.push(3)
stack.push(9)
stack.push(6)

print("Original Stack:", stack.items)
stack.reverse()
print("Reversed Stack:", stack.items)
```

```python
#You are given a string S, the task is to reverse the string using stack.
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)


def reverse_string(string):
    stack = Stack()

    # Push each character of the string onto the stack
    for char in string:
        stack.push(char)

    reversed_string = ""

    # Pop the characters from the stack to obtain the reversed string
    while not stack.is_empty():
        reversed_string += stack.pop()

    return reversed_string


# Example usage:
input_string = "GeeksforGeeks"
reversed_string = reverse_string(input_string)
print(reversed_string)
```

```python
#Given string S representing a postfix expression, the task is to evaluate the expression and find the final value. Operators will only include the basic arithmetic operators like *, /, + and -.

class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)


def evaluate_postfix(expression):
    stack = Stack()

    # Iterate through each character in the expression
    for char in expression:
        if char.isdigit():
            # If the character is a digit, push it onto the stack
            stack.push(int(char))
        else:
            # If the character is an operator, pop two operands from the stack
            operand2 = stack.pop()
            operand1 = stack.pop()

            # Perform the operation based on the operator and push the result onto the stack
            if char == '*':
                result = operand1 * operand2
            elif char == '/':
                result = operand1 / operand2
            elif char == '+':
                result = operand1 + operand2
            elif char == '-':
                result = operand1 - operand2

            stack.push(result)

    # The final result will be the only remaining element in the stack
    return stack.pop()


# Example usage:
expression = "231*+9-"
result = evaluate_postfix(expression)
print(result)

expression = "123+*8-"
result = evaluate_postfix(expression)
print(result)


```

```python

#Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.
class MinStack:
    def __init__(self):
        self.stack = []
        self.min_stack = []

    def push(self, val):
        self.stack.append(val)
        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)

    def pop(self):
        if self.stack:
            popped_element = self.stack.pop()
            if popped_element == self.min_stack[-1]:
                self.min_stack.pop()

    def top(self):
        if self.stack:
            return self.stack[-1]

    def getMin(self):
        if self.min_stack:
            return self.min_stack[-1]


# Example usage:
minStack = MinStack()
minStack.push(-2)
minStack.push(0)
minStack.push(-3)
print(minStack.getMin())  # Output: -3
minStack.pop()
print(minStack.top())     # Output: 0
print(minStack.getMin())  # Output: -2
```

```python
#Given n non-negative integers representing an elevation map where the width of each bar is 1, compute how much water it can trap after raining.

def trap_water(height):
    if not height:
        return 0

    n = len(height)
    left = 0
    right = n - 1
    left_max = 0
    right_max = 0
    water = 0

    while left < right:
        if height[left] < height[right]:
            if height[left] > left_max:
                left_max = height[left]
            else:
                water += left_max - height[left]
            left += 1
        else:
            if height[right] > right_max:
                right_max = height[right]
            else:
                water += right_max - height[right]
            right -= 1

    return water


# Example usage:
height = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
water_trapped = trap_water(height)
print(water_trapped)  # Output: 6

height = [4, 2, 0, 3, 2, 5]
water_trapped = trap_water(height)
print(water_trapped)  # Output: 9
```
