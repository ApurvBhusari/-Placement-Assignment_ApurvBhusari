```python
# Definition for a binary tree node
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def findMaxSubtreeSum(self, root):
        # Helper function to calculate the sum of a subtree rooted at a given node
        def subtreeSum(node):
            if not node:
                return 0
            
            left_sum = subtreeSum(node.left)
            right_sum = subtreeSum(node.right)
            
            return node.val + left_sum + right_sum
        
        # Function to find the maximum sum subtree
        def maxSubtreeSum(node, max_sum):
            if not node:
                return 0
            
            left_sum = maxSubtreeSum(node.left, max_sum)
            right_sum = maxSubtreeSum(node.right, max_sum)
            
            current_sum = node.val + left_sum + right_sum
            
            max_sum[0] = max(max_sum[0], current_sum)
            
            return current_sum if current_sum > 0 else 0
        
        # Main function
        if not root:
            return 0
        
        max_sum = [float('-inf')]
        maxSubtreeSum(root, max_sum)
        
        return max_sum[0]

# Test case 1
root1 = TreeNode(1)
root1.left = TreeNode(2)
root1.right = TreeNode(3)
root1.left.left = TreeNode(4)
root1.left.right = TreeNode(5)
root1.right.left = TreeNode(6)
root1.right.right = TreeNode(7)

solution = Solution()
print(solution.findMaxSubtreeSum(root1))  # Output: 28

# Test case 2
root2 = TreeNode(1)
root2.left = TreeNode(-2)
root2.right = TreeNode(3)
root2.left.left = TreeNode(4)
root2.left.right = TreeNode(5)
root2.right.left = TreeNode(-6)
root2.right.right = TreeNode(2)

print(solution.findMaxSubtreeSum(root2))  # Output: 7
```

```python
# Definition for a binary tree node
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def constructBST(self, arr):
        # Helper function to construct a BST from a given level order traversal
        def constructBSTUtil(arr, start, end):
            if start > end:
                return None
            
            mid = (start + end) // 2
            node = TreeNode(arr[mid])
            
            node.left = constructBSTUtil(arr, start, mid - 1)
            node.right = constructBSTUtil(arr, mid + 1, end)
            
            return node
        
        # Main function
        if not arr:
            return None
        
        return constructBSTUtil(arr, 0, len(arr) - 1)

# Test case
arr = [7, 4, 12, 3, 6, 8, 1, 5, 10]

solution = Solution()
root = solution.constructBST(arr)

# Function to print the BST level by level
def printLevelOrder(root):
    if not root:
        return
    
    queue = [root]
    
    while queue:
        node = queue.pop(0)
        print(node.val, end=" ")
        
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    
    print()

printLevelOrder(root)
```

```python
def canRepresentBST(arr):
    if not arr:
        return True

    n = len(arr)
    bounds = [(float('-inf'), float('inf'))]

    for i in range(1, n):
        node = arr[i]
        parent_lower, parent_upper = bounds[i - 1]

        if node <= parent_lower or node >= parent_upper:
            return False

        if i < n - 1:
            if node < arr[i + 1]:
                bounds.append((parent_lower, node))
            else:
                bounds.append((node, parent_upper))

    return True

# Test case 1
arr1 = [7, 4, 12, 3, 6, 8, 1, 5, 10]
print(canRepresentBST(arr1))  # Output: True

# Test case 2
arr2 = [11, 6, 13, 5, 12, 10]
print(canRepresentBST(arr2))  # Output: False
```
