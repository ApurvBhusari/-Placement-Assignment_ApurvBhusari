```python
#1


#Given two linked list of the same size, the task is to create a new linked list using those linked lists. The condition is that the greater node among both linked list will be added to the new linked list

class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None


def create_new_linked_list(list1, list2):
    if not list1 or not list2:  # If any of the lists is empty, return None
        return None

    # Initialize pointers for both lists
    current1 = list1
    current2 = list2

    # Create a dummy node as the head of the new list
    new_list_head = Node()
    current_new = new_list_head

    while current1 and current2:
        if current1.data >= current2.data:
            current_new.next = Node(current1.data)
            current1 = current1.next
        else:
            current_new.next = Node(current2.data)
            current2 = current2.next

        current_new = current_new.next

    return new_list_head.next  # Return the head of the new list


# Helper function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.next
    print()


# Test case 1
# Create the first linked list: 5->2->3->8
list1 = Node(5)
list1.next = Node(2)
list1.next.next = Node(3)
list1.next.next.next = Node(8)

# Create the second linked list: 1->7->4->5
list2 = Node(1)
list2.next = Node(7)
list2.next.next = Node(4)
list2.next.next.next = Node(5)

# Create the new linked list using the given lists
new_list = create_new_linked_list(list1, list2)

# Print the new linked list
print("New list:")
print_linked_list(new_list)

# Test case 2
# Create the first linked list: 2->8->9->3
list1 = Node(2)
list1.next = Node(8)
list1.next.next = Node(9)
list1.next.next.next = Node(3)

# Create the second linked list: 5->3->6->4
list2 = Node(5)
list2.next = Node(3)
list2.next.next = Node(6)
list2.next.next.next = Node(4)

# Create the new linked list using the given lists
new_list = create_new_linked_list(list1, list2)

# Print the new linked list
print("New list:")
print_linked_list(new_list)
```

```python
#2
#Write a function that takes a list sorted in non-decreasing order and deletes any duplicate nodes from the list. The list should only be traversed once.


class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None


def removeDuplicates(head):
    if not head or not head.next:  # If the list is empty or contains only one node
        return head

    current = head
    while current.next:
        if current.data == current.next.data:
            current.next = current.next.next  # Skip the next node
        else:
            current = current.next  # Move to the next node

    return head


# Helper function to print the linked list
def printLinkedList(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.next
    print()


# Test case 1
# Create the linked list: 11->11->11->21->43->43->60
head1 = Node(11)
head1.next = Node(11)
head1.next.next = Node(11)
head1.next.next.next = Node(21)
head1.next.next.next.next = Node(43)
head1.next.next.next.next.next = Node(43)
head1.next.next.next.next.next.next = Node(60)

print("Input:")
printLinkedList(head1)

# Remove duplicates
newHead1 = removeDuplicates(head1)

print("Output:")
printLinkedList(newHead1)


# Test case 2
# Create the linked list: 10->12->12->25->25->25->34
head2 = Node(10)
head2.next = Node(12)
head2.next.next = Node(12)
head2.next.next.next = Node(25)
head2.next.next.next.next = Node(25)
head2.next.next.next.next.next = Node(25)
head2.next.next.next.next.next.next = Node(34)

print("Input:")
printLinkedList(head2)

# Remove duplicates
newHead2 = removeDuplicates(head2)

print("Output:")
printLinkedList(newHead2)

```

```python
#3
#Given a linked list of size N. The task is to reverse every k nodes (where k is an input to the function) in the linked list. If the number of nodes is not a multiple of k then left-out nodes, in the end, should be considered as a group and must be reversed (See Example 2 for clarification).

class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None


def reverseKNodes(head, k):
    if not head or k == 1:  # If the list is empty or k is 1, no reversal needed
        return head

    dummy = Node()  # Create a dummy node as the head of the new list
    dummy.next = head

    prev = dummy
    current = head
    count = 0

    while current:
        count += 1

        if count % k == 0:  # Reached a group of k nodes
            prev = reverseGroup(prev, current.next)
            current = prev.next
        else:
            current = current.next

    return dummy.next


def reverseGroup(prev, next_node):
    last = prev.next
    current = last.next

    while current != next_node:
        last.next = current.next
        current.next = prev.next
        prev.next = current
        current = last.next

    return last


# Helper function to print the linked list
def printLinkedList(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.next
    print()


# Test case 1
# Create the linked list: 1->2->2->4->5->6->7->8
head1 = Node(1)
head1.next = Node(2)
head1.next.next = Node(2)
head1.next.next.next = Node(4)
head1.next.next.next.next = Node(5)
head1.next.next.next.next.next = Node(6)
head1.next.next.next.next.next.next = Node(7)
head1.next.next.next.next.next.next.next = Node(8)

k1 = 4

print("Input:")
printLinkedList(head1)

# Reverse every k nodes
newHead1 = reverseKNodes(head1, k1)

print("Output:")
printLinkedList(newHead1)


# Test case 2
# Create the linked list: 1->2->3->4->5
head2 = Node(1)
head2.next = Node(2)
head2.next.next = Node(3)
head2.next.next.next = Node(4)
head2.next.next.next.next = Node(5)

k2 = 3

print("Input:")
printLinkedList(head2)

# Reverse every k nodes
newHead2 = reverseKNodes(head2, k2)

print("Output:")
printLinkedList(newHead2)

```

```python
#5  Given a linked list and a key to be deleted. Delete last occurrence of key from linked. The list may have duplicates.

class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None


def deleteLastOccurrence(head, key):
    if not head:
        return head

    prev = None
    last = None
    toDelete = None

    current = head

    # Traverse the linked list to find the last occurrence of the key
    while current:
        if current.data == key:
            last = prev
            toDelete = current
        prev = current
        current = current.next

    # If the last occurrence is at the beginning of the list, update the head
    if toDelete == head:
        head = head.next
    else:
        last.next = toDelete.next

    # Delete the node
    toDelete.next = None

    return head


# Helper function to print the linked list
def printLinkedList(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.next
    print()


# Test case
# Create the linked list: 1->2->3->5->2->10
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(5)
head.next.next.next.next = Node(2)
head.next.next.next.next.next = Node(10)

key = 2

print("Input:")
printLinkedList(head)

# Delete the last occurrence of the key
newHead = deleteLastOccurrence(head, key)

print("Output:")
printLinkedList(newHead)
```

```python
#6 Given two sorted linked lists consisting of N and M nodes respectively. The task is to merge both of the lists (in place) and return the head of the merged list.

class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None


def mergeSortedLists(a, b):
    # Create a dummy node as the head of the merged list
    dummy = Node()
    result = dummy

    # Merge the lists until either 'a' or 'b' becomes None
    while a and b:
        if a.data <= b.data:
            result.next = a
            a = a.next
        else:
            result.next = b
            b = b.next
        result = result.next

    # Append the remaining nodes from 'a' or 'b' to the merged list
    if a:
        result.next = a
    else:
        result.next = b

    # Return the head of the merged list
    return dummy.next


# Helper function to print the linked list
def printLinkedList(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.next
    print()


# Test cases
# Create the linked list a: 5->10->15
a = Node(5)
a.next = Node(10)
a.next.next = Node(15)




b = Node(2)
b.next = Node(3)
b.next.next = Node(20)

print("Input:")
print("a:", end=" ")
printLinkedList(a)
print("b:", end=" ")
printLinkedList(b)
merged = mergeSortedLists(a, b)

print("Output:")
printLinkedList(merged)

# Create the linked list a: 1->1
a = Node(1)
a.next = Node(1)

# Create the linked list b: 2->4
b = Node(2)
b.next = Node(4)

print("Input:")
print("a:", end=" ")
printLinkedList(a)
print("b:", end=" ")
printLinkedList(b)

# Merge the two sorted lists
merged = mergeSortedLists(a, b)

print("Output:")
printLinkedList(merged)

```

```python
#7 Given a Doubly Linked List, the task is to reverse the given Doubly Linked List.

class Node:
    def __init__(self, data=None):
        self.data = data
        self.prev = None
        self.next = None


def reverseDoublyLinkedList(head):
    if not head or not head.next:  # If the list is empty or has only one node, no reversal needed
        return head

    current = head
    prev = None

    # Traverse the linked list and swap the next and prev pointers
    while current:
        next_node = current.next
        current.next = prev
        current.prev = next_node
        prev = current
        current = next_node

    # Update the head pointer to the last node (new head of the reversed list)
    head = prev

    return head


# Helper function to print the doubly linked list
def printDoublyLinkedList(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.next
    print()


# Test case
# Create the doubly linked list: 10<->8<->4<->2
head = Node(10)
node2 = Node(8)
node3 = Node(4)
node4 = Node(2)

head.next = node2
node2.prev = head
node2.next = node3
node3.prev = node2
node3.next = node4
node4.prev = node3

print("Original Linked list:")
printDoublyLinkedList(head)

# Reverse the doubly linked list
reversed_head = reverseDoublyLinkedList(head)

print("Reversed Linked list:")
printDoublyLinkedList(reversed_head)
```

```python
#8  Given a doubly linked list and a position. The task is to delete a node from given position in a doubly linked list.

class Node:
    def __init__(self, data=None):
        self.data = data
        self.prev = None
        self.next = None


def deleteNode(head, position):
    if not head:  # If the list is empty, nothing to delete
        return head

    if position == 1:  # Delete the head node
        new_head = head.next
        if new_head:  # If there is a new head, update its prev pointer
            new_head.prev = None
        # Delete the current head node (optional, depending on the programming language)
        del head
        return new_head

    current = head
    count = 1

    # Traverse the list to find the node at the given position
    while current and count < position:
        current = current.next
        count += 1

    if not current:  # If the position exceeds the length of the list, return the original head
        return head

    prev_node = current.prev
    next_node = current.next

    if prev_node:  # Update the next pointer of the previous node
        prev_node.next = next_node

    if next_node:  # Update the prev pointer of the next node
        next_node.prev = prev_node

    # Delete the current node (optional, depending on the programming language)
    del current

    return head


# Helper function to print the doubly linked list
def printDoublyLinkedList(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.next
    print()


# Test case 1
# Create the doubly linked list: 1<->3<->4
head1 = Node(1)
node2 = Node(3)
node3 = Node(4)

head1.next = node2
node2.prev = head1
node2.next = node3
node3.prev = node2

print("Input:")
printDoublyLinkedList(head1)

# Delete node at position 3
position1 = 3
updated_head1 = deleteNode(head1, position1)

print("Output:")
printDoublyLinkedList(updated_head1)


# Test case 2
# Create the doubly linked list: 1<->5<->2<->9
head2 = Node(1)
node2 = Node(5)
node3 = Node(2)
node4 = Node(9)

head2.next = node2
node2.prev = head2
node2.next = node3
node3.prev = node2
node3.next = node4
node4.prev = node3

print("\nInput:")
printDoublyLinkedList(head2)

# Delete node at position 1
position2 = 1
updated_head2 = deleteNode(head2, position2)

print("Output:")
printDoublyLinkedList(updated_head2)
```
