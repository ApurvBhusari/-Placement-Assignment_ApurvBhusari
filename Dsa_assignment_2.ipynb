```python
#Given an integer array nums of 2n integers, group these integers into n pairs (a1, b1), (a2, b2),..., (an, bn) such that the sum of min(ai, bi) for all i is maximized. Return the maximized sum.

#Example 1:
#Input: nums = [1,4,3,2]
#Output: 4

class ArrayPairSum:
    def __init__(self, nums):
        self.nums = nums

    def bubble_sort(self):
        n = len(self.nums)
        for i in range(n - 1):
            for j in range(n - i - 1):
                if self.nums[j] > self.nums[j + 1]:
                    self.nums[j], self.nums[j + 1] = self.nums[j + 1], self.nums[j]

    def find_max_sum(self):
        self.bubble_sort()  # Sort the array using bubble sort
        total_sum = 0
        if len(self.nums)%2==0:
            for i in range(0, len(self.nums), 2):
                total_sum += self.nums[i]
            return total_sum
        elif len(self.nums)%2!=0:
            for i in range(1,len(self.nums),2):
                total_sum+=self.nums[i]
            return total_sum
        
        


#1
nums = [1, 4, 3, 2,5]
pair_sum = ArrayPairSum(nums)
max_sum = pair_sum.find_max_sum()
print(max_sum)  # Output: 4
```

```python
#Question 2
#Alice has n candies, where the ith candy is of type candyType[i]. Alice noticed that she started to gain weight, so she visited a doctor. 

#The doctor advised Alice to only eat n / 2 of the candies she has (n is always even). Alice likes her candies very much, and she wants to eat the maximum number of different types of candies while still following the doctor's advice. 

#Given the integer array candyType of length n, return the maximum number of different types of candies she can eat if she only eats n / 2 of them.

#Example 1:
#Input: candyType = [1,1,2,2,3,3]
#Output: 3

#Explanation: Alice can only eat 6 / 2 = 3 candies. Since there are only 3 types, she can eat one of each type.

class CandyCounter:
    def __init__(self, candyType):
        self.candyType = candyType
    
    def countUniqueCandies(self):
        unique_candies = 0
        seen = []
        for candy in self.candyType:
            if candy not in seen:
                unique_candies += 1
                seen.append(candy)
        return unique_candies
    
    def maxCandies(self):
        unique_candies = self.countUniqueCandies()
        max_candies = len(self.candyType) // 2
        return min(unique_candies, max_candies)

```

```python
#Question 3
#We define a harmonious array as an array where the difference between its maximum value
#and its minimum value is exactly 1.

#Given an integer array nums, return the length of its longest harmonious subsequence
#among all its possible subsequences.

#A subsequence of an array is a sequence that can be derived from the array by deleting some or no elements without changing the order of the remaining elements.

#Example 1:
#Input: nums = [1,3,2,2,5,2,3,7]
#Output: 5

#Explanation: The longest harmonious subsequence is [3,2,2,2,3]


class HarmoniousSubsequence:
    def __init__(self, nums):
        self.nums = nums
    
    def findLHS(self):
        num_freq = self.calculateFrequency()
        
        max_length = 0
        for key in num_freq.keys():
            if key + 1 in num_freq:
                length = num_freq[key] + num_freq[key + 1]
                max_length = max(max_length, length)
        
        return max_length
    
    def calculateFrequency(self):
        num_freq = {}
        for num in self.nums:
            num_freq[num] = num_freq.get(num, 0) + 1
        
        return num_freq

nums = [1, 3, 2, 2, 5, 2, 3, 7]
hs = HarmoniousSubsequence(nums)
print(hs.findLHS())

nums1 = [1,3,2,2,3,4,5,5,6,7,4,5,4,5,5]
hs = HarmoniousSubsequence(nums1)
print(hs.findLHS())

nums2=[1,3,2,2,3,4,5,2,3,5,4]
hs=HarmoniousSubsequence(nums2)
print(hs.findLHS())




```

```python
#4You have a long flowerbed in which some of the plots are planted, and some are not.
#However, flowers cannot be planted in adjacent plots.
#Given an integer array flowerbed containing 0's and 1's, where 0 means empty and 1 means not empty, and an integer n, return true if n new flowers can be planted in the flowerbed without violating the no-adjacent-flowers rule and false otherwise.

#Example 1:
#Input: flowerbed = [1,0,0,0,1], n = 1
#Output: true

class FlowerBed:
    def __init__(self, flowerbed):
        self.flowerbed = flowerbed
    
    def canPlaceFlowers(self, n):
        length = len(self.flowerbed)
        count = 0
        for i in range(length):
            if (
                self.flowerbed[i] == 0 and 
                (i == 0 or self.flowerbed[i - 1] == 0) and 
                (i == length - 1 or self.flowerbed[i + 1] == 0)
            ):
                self.flowerbed[i] = 1
                count += 1
            if count >= n:
                return True
        return False
flowerbed = [1, 0, 0, 0, 1]
n = 1
fb = FlowerBed(flowerbed)
print(fb.canPlaceFlowers(n))

flowerbed1 = [1, 0, 0, 0,0, 1]
n = 2
fb = FlowerBed(flowerbed1)
print(fb.canPlaceFlowers(n))

flowerbed2 = [1, 0, 0, 0,0,0, 1]
n = 2
fb = FlowerBed(flowerbed2)
print(fb.canPlaceFlowers(n))

```

```python
#5Given an integer array nums, find three numbers whose product is maximum and return the maximum product.
#Example 1:
#Input: nums = [1,2,3]
#Output: 6
class MaxProduct:
    def __init__(self, nums):
        self.nums = nums
    
    def maximumProduct(self):
        max1 = max2 = max3 = float('-inf')
        min1 = min2 = float('inf')

        for num in self.nums:
            if num > max1:
                max1, max2, max3 = num, max1, max2
            elif num > max2:
                max2, max3 = num, max2
            elif num > max3:
                max3 = num

            if num < min1:
                min1, min2 = num, min1
            elif num < min2:
                min2 = num

        return max(max1 * max2 * max3, min1 * min2 * max1)
    
nums = [1, 2, 3]
mp = MaxProduct(nums)
print(mp.maximumProduct())
        
nums2 = [5,6,7,3,9,10,1,2,3,4,66,67,90,0,99,43,76,4]
mp = MaxProduct(nums2)
print(mp.maximumProduct())

```

```python
#6Given an array of integers nums which is sorted in ascending order, and an integer target,
#write a function to search target in nums. If target exists, then return its index. Otherwise,
#return -1.

#You must write an algorithm with O(log n) runtime complexity.

#Input: nums = [-1,0,3,5,9,12], target = 9
#Output: 4

#Explanation: 9 exists in nums and its index is 4

class BinarySearch:
    def __init__(self, nums):
        self.nums = nums
    
    def search(self, target):
        left, right = 0, len(self.nums) - 1

        while left <= right:
            mid = (left + right) // 2
            if self.nums[mid] == target:
                return mid
            elif self.nums[mid] > target:
                right = mid - 1
            else:
                left = mid + 1

        return -1
nums = [-1, 0, 3, 5, 9, 12]
target = 9
bs = BinarySearch(nums)
print(bs.search(target))

nums1=[1,2,5,6,-1,55,32]
target1=5
bs1=BinarySearch(nums1)
print(bs1.search(target1))
```

```python
#7
#An array is monotonic if it is either monotone increasing or monotone decreasing.

#An array nums is monotone increasing if for all i <= j, nums[i] <= nums[j]. An array nums is
#monotone decreasing if for all i <= j, nums[i] >= nums[j].

#Given an integer array nums, return true if the given array is monotonic, or false otherwise.

#Example 1:
#Input: nums = [1,2,2,3]
#Output: true


class MonotonicArray:d
    def __init__(self, nums):
        self.nums = nums
    
    def isMonotonic(self):
        isIncreasing = True
        isDecreasing = True

        for i in range(1, len(self.nums)):
            if self.nums[i] < self.nums[i-1]:
                isIncreasing = False
            if self.nums[i] > self.nums[i-1]:
                isDecreasing = False

        return isIncreasing or isDecreasing
nums1 = [1, 2, 2, 3]
ma = MonotonicArray(nums1)
print(ma.isMonotonic())


nums2 = [1, 2, 2, 3,2]
ma = MonotonicArray(nums2)
print(ma.isMonotonic())


nums3 = [4,3,3,3,2,2,2,1]
ma = MonotonicArray(nums3)
print(ma.isMonotonic())

```

```python
#8
#You are given an integer array nums and an integer k.

#In one operation, you can choose any index i where 0 <= i < nums.length and change nums[i] to nums[i] + x where x is an integer from the range [-k, k]. You can apply this operation at most once for each index i.

#The score of nums is the difference between the maximum and minimum elements in nums.

#Return the minimum score of nums after applying the mentioned operation at most once for each index in it.

#Example 1:
#Input: nums = [1], k = 0
#Output: 0

#Explanation: The score is max(nums) - min(nums) = 1 - 1 = 0.


class MinimumScore:
    def __init__(self, nums, k):
        self.nums = nums
        self.k = k
    
    def calculateScore(self):
        min_num = min(self.nums)
        max_num = max(self.nums)

        if max_num - min_num <= 2 * self.k:
            return 0

        average = (min_num + max_num) // 2
        min_num = average - self.k
        max_num = average + self.k

        for i in range(len(self.nums)):
            if self.nums[i] < min_num:
                self.nums[i] = min_num
            elif self.nums[i] > max_num:
                self.nums[i] = max_num

        return max_num - min_num
    
nums1 = [1]
k1 = 0
ms = MinimumScore(nums1, k1)
print(ms.calculateScore())

nums2 = [1,2,3]
k2 = 5
ms = MinimumScore(nums2, k2)
print(ms.calculateScore())

nums3 = [1,1,2,5]
k3 = 2
ms = MinimumScore(nums3, k3)
print(ms.calculateScore())

```
