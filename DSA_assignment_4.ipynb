```python
#1
#Given three integer arrays arr1, arr2 and arr3 **sorted** in **strictly increasing** order, return a sorted array of **only** the integers that appeared in **all** three arrays.

#**Example 1:**

#Input: arr1 = [1,2,3,4,5], arr2 = [1,2,5,7,9], arr3 = [1,3,4,5,8]

#Output: [1,5]

#**Explanation:** Only 1 and 5 appeared in the three arrays.

from typing import List

class CommonElementsFinder:
    def __init__(self, arr1: List[int], arr2: List[int], arr3: List[int]):
        self.arr1 = arr1
        self.arr2 = arr2
        self.arr3 = arr3

    def find_common_elements(self) -> List[int]:
        set1 = set(self.arr1)
        set2 = set(self.arr2)
        set3 = set(self.arr3)

        common_elements = set1.intersection(set2, set3)

        return sorted(list(common_elements))


```

```python
#2

#Given two **0-indexed** integer arrays nums1 and nums2, return *a list* answer *of size* 2 *where:*

#- answer[0] *is a list of all **distinct** integers in* nums1 *which are **not** present in* nums2*.*
#- answer[1] *is a list of all **distinct** integers in* nums2 *which are **not** present in* nums1.

#**Note** that the integers in the lists may be returned in **any** order.

#**Example 1:**

#**Input:** nums1 = [1,2,3], nums2 = [2,4,6]

#**Output:** [[1,3],[4,6]]

#**Explanation:**

#For nums1, nums1[1] = 2 is present at index 0 of nums2, whereas nums1[0] = 1 and nums1[2] = 3 are not present in nums2. Therefore, answer[0] = [1,3].

#For nums2, nums2[0] = 2 is present at index 1 of nums1, whereas nums2[1] = 4 and nums2[2] = 6 are not present in nums2. Therefore, answer[1] = [4,6].

from typing import List

class DistinctNumbersFinder:
    def __init__(self, nums1: List[int], nums2: List[int]):
        self.nums1 = nums1
        self.nums2 = nums2

    def find_distinct_numbers(self) -> List[List[int]]:
        set1 = set(self.nums1)
        set2 = set(self.nums2)

        list1 = list(set1 - set2)
        list2 = list(set2 - set1)

        return [list1, list2]
nums1 = [1, 2, 3]
nums2 = [2, 4, 6]

finder = DistinctNumbersFinder(nums1, nums2)
result = finder.find_distinct_numbers()
print(result)

nums3 = [4,6,9,15]
nums4 = [20,4,6,3,99,47]

finder1 = DistinctNumbersFinder(nums3, nums4)
result1 = finder1.find_distinct_numbers()
print(result1)


```

```python
#3 Given a 2D integer array matrix, return *the **transpose** of* matrix.

#The **transpose** of a matrix is the matrix flipped over its main diagonal, switching the matrix's row and column indices.

#**Example 1:**

#Input: matrix = [[1,2,3],[4,5,6],[7,8,9]]

#Output: [[1,4,7],[2,5,8],[3,6,9]]

from typing import List

class MatrixTransposer:
    def __init__(self, matrix: List[List[int]]):
        self.matrix = matrix

    def transpose(self) -> List[List[int]]:
        rows = len(self.matrix)
        cols = len(self.matrix[0])
        print(rows)
        print(cols)

        # Create a new matrix with swapped dimensions
        transposed = [[0 for a in range(rows)] for a in range(cols)]

        # Perform the transpose operation
        for i in range(rows):
            for j in range(cols):
                transposed[j][i] = self.matrix[i][j]

        return transposed
    
    
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
transposer = MatrixTransposer(matrix)
result = transposer.transpose()
print(result)


matrix1 = [[1, 4,2], [15,14,10], [65,23,8]]
transposer1 = MatrixTransposer(matrix1)
result1 = transposer1.transpose()
print(result1)
```

```python
#4
#Given an integer array nums of 2n integers, group these integers into n pairs (a1, b1), (a2, b2), ..., (an, bn) such that the sum of min(ai, bi) for all i is **maximized**. Return *the maximized sum*.

#**Example 1:**

#Input: nums = [1,4,3,2]

#Output: 4

#**Explanation:** All possible pairings (ignoring the ordering of elements) are:

#1. (1, 4), (2, 3) -> min(1, 4) + min(2, 3) = 1 + 2 = 3

#2. (1, 3), (2, 4) -> min(1, 3) + min(2, 4) = 1 + 2 = 3

#3. (1, 2), (3, 4) -> min(1, 2) + min(3, 4) = 1 + 3 = 4

#So the maximum possible sum is 4.

def arrayPairSum(nums):
    # Find the maximum and minimum values in the array
    max_val = max(nums)
    min_val = min(nums)

    # Calculate the range of values
    value_range = max_val - min_val + 1

    # Create a count array to store the occurrences of each number
    count = [0] * value_range

    # Count the occurrences of each number
    for num in nums:
        count[num - min_val] += 1

    # Iterate over the count array to construct the pairs and calculate the maximum sum
    pair_sum = 0
    take = True  # Flag to indicate whether to include the number in the sum
    for i in range(value_range):
        while count[i] > 0:
            if take:
                pair_sum += i + min_val
            take = not take
            count[i] -= 1

    return pair_sum


nums = [1, 4, 3, 2]
result = arrayPairSum(nums)
print(result)

nums1 = [1, 5,4,9]
result1 = arrayPairSum(nums1)
print(result1)


```

```python
#5
#You have n coins and you want to build a staircase with these coins. The staircase consists of k rows where the ith row has exactly i coins. The last row of the staircase **may be** incomplete.

#Given the integer n, return *the number of **complete rows** of the staircase you will build*.

#**Example 1:**

#[]()

#![v2.jpg](https://s3-us-west-2.amazonaws.com/secure.notion-static.com/4bd91cfa-d2b1-47b3-8197-a72e8dcfff4b/v2.jpg)

#**Input:** n = 5

#**Output:** 2

#**Explanation:** Because the 3rd row is incomplete, we return 2.


class Solution:
    def arrangeCoins(self, n: int) -> int:
        left = 0
        right = n
        
        while left <= right:
            mid = left + (right - left) // 2
            curr = mid * (mid + 1) // 2
            
            if curr == n:
                return mid
            elif curr < n:
                left = mid + 1
            else:
                right = mid - 1
        
        return right

n = 5
solution = Solution()
result = solution.arrangeCoins(n)
print(result)


n1 = 9
solution1 = Solution()
result1 = solution1.arrangeCoins(n1)
print(result1)


n2=15
solution2=Solution()
result2=solution2.arrangeCoins(n2)
print(result2)
```

```python
#6
#Given an integer array nums sorted in **non-decreasing** order, return *an array of **the squares of each number** sorted in non-decreasing order*.

#**Example 1:**

#Input: nums = [-4,-1,0,3,10]

#Output: [0,1,9,16,100]

#**Explanation:** After squaring, the array becomes [16,1,0,9,100].
#After sorting, it becomes [0,1,9,16,100]

from typing import List

class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        n = len(nums)
        result = [0] * n
        left, right = 0, n - 1
        index = n - 1

        while left <= right:
            if abs(nums[left]) > abs(nums[right]):
                result[index] = nums[left] * nums[left]
                left += 1
            else:
                result[index] = nums[right] * nums[right]
                right -= 1
            index -= 1

        return result
nums = [-4, -1, 0, 3, 10]
solution = Solution()
result = solution.sortedSquares(nums)
print(result)

nums1 = [-4, -1, 0, 3, 10]
solution1 = Solution()
result1 = solution1.sortedSquares(nums1)
print(result1)
```

```python
#7

#You are given an m x n matrix M initialized with all 0's and an array of operations ops, where ops[i] = [ai, bi] means M[x][y] should be incremented by one for all 0 <= x < ai and 0 <= y < bi.

#Count and return *the number of maximum integers in the matrix after performing all the operations*

#**Example 1:**

#![q4.jpg](https://s3-us-west-2.amazonaws.com/secure.notion-static.com/4d0890d0-7bc7-4f59-be8e-352d9f3c1c52/q4.jpg)

#**Input:** m = 3, n = 3, ops = [[2,2],[3,3]]

#**Output:** 4

#**Explanation:** The maximum integer in M is 2, and there are four of it in M. So return 4.

class Solution:
    def maxCount(self, m: int, n: int, ops: List[List[int]]) -> int:
        if not ops:
            return m * n
        
        min_a = float('inf')
        min_b = float('inf')
        
        for op in ops:
            min_a = min(min_a, op[0])
            min_b = min(min_b, op[1])
        
        return min_a * min_b
m = 3
n = 3
ops = [[2,2],[3,4]]

solution = Solution()
result = solution.maxCount(m, n, ops)
print(result)


m1 = 3
n1 = 3
ops1 = [[1,2],[3,3]]

solution1 = Solution()
result1 = solution1.maxCount(m1, n1, ops1)
print(result1)


```

```python
#8
#Given the array nums consisting of 2n elements in the form [x1,x2,...,xn,y1,y2,...,yn].

#*Return the array in the form* [x1,y1,x2,y2,...,xn,yn].

#**Example 1:**

#**Input:** nums = [2,5,1,3,4,7], n = 3

#**Output:** [2,3,5,4,1,7]

#**Explanation:** Since x1=2, x2=5, x3=1, y1=3, y2=4, y3=7 then the answer is [2,3,5,4,1,7].

from typing import List

class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        result = []
        for i in range(n):
            result.append(nums[i])
            result.append(nums[n + i])
        return result
nums = [2, 5, 1, 3, 4, 7]
n = 2
solution = Solution()
result = solution.shuffle(nums, n)
print(result)


nums1 = [2, 5, 1, 3, 4, 7,-3,9]
n1 = 4
solution1 = Solution()
result1 = solution1.shuffle(nums1, n1)
print(result1)
```
